```python
import pandas as pd 
import matplotlib as plt 
import numpy as np
```

```python
video = pd.read_csv("USvideos.csv")
```

```python
video.head(10)
```

```python
video["ratio"]= video["likes"]/video["views"]
```

```python
video.head(20)
```

```python
out=[]
def Zscore_outlier(df):
    m = np.mean(df)
    sd = np.std(df)
    for i in df: 
        z = (i-m)/sd
        if np.abs(z) > 3: 
            out.append(i)
    print("Outliers:",out)
Zscore_outlier(video['ratio'])
```

```python
out2=[]
def Winsorization_outliers(video):
    q1 = np.percentile(video , 1)
    q3 = np.percentile(video , 99)
    for i in video:
        if i > q3 or i < q1:
            out2.append(i)
    print("Outliers:",out2)
Winsorization_outliers(video['ratio'])
```

```python
video.info()
```

```python
video = video[~video['ratio'].isin(out2)]
```

```python
video.info()
```

```python
video.to_csv('USvideos.csv', index=False)
```
