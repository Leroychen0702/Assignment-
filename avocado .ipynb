```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("avocado_cleaned (1).csv") 
```

```python
df.head(10)
```

```python


df['year'] = df['Date'].dt.year
```

```python
df.head()
```

```python



aaa = df.groupby("year")["TotalVolume"].sum().reset_index()
bbb = df.groupby("year")["plu4046"].sum().reset_index()
ccc = df.groupby("year")["plu4225"].sum().reset_index()
ddd = df.groupby("year")["plu4770"].sum().reset_index()


fig, ax = plt.subplots()


aaa.plot(x='year', y='TotalVolume', kind='line', label='Series 1', ax=ax)
bbb.plot(x='year', y='plu4046', kind='line', label='Series 2', ax=ax)
ccc.plot(x='year', y='plu4225', kind='line', label='Series 3', ax=ax)
ddd.plot(x='year', y='plu4770', kind='line', label='Series 4', ax=ax)


plt.xlabel("Year")
plt.ylabel("Volume")
plt.legend()


plt.show()
```

```python
# Assuming you have a DataFrame 'df' with columns 'year', 'TotalVolume', 'plu4046', 'plu4225', and 'plu4770'
# df = ...

# Group and aggregate the data for each series

bbb = df.groupby("year")["plu4046"].sum().reset_index()
ccc = df.groupby("year")["plu4225"].sum().reset_index()
ddd = df.groupby("year")["plu4770"].sum().reset_index()

# Create a single figure and axis
fig, ax = plt.subplots()

# Plot each series on the same axis

bbb.plot(x='year', y='plu4046', kind='line', label='Series 1', ax=ax)
ccc.plot(x='year', y='plu4225', kind='line', label='Series 2', ax=ax)
ddd.plot(x='year', y='plu4770', kind='line', label='Series 3', ax=ax)

# Set labels and legend
plt.xlabel("Year")
plt.ylabel("Volume")
plt.legend()

# Show the plot
plt.show()
```
