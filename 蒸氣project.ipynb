```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
genre = pd.read_csv("US_category_id.csv")
video = pd.read_csv("USvideos.csv")
```

```python
genre.head()
```

```python
video.head()
```

```python
df = pd.merge( video, genre, how= 'inner', left_on='category_id', right_on='id' )
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df["snippet/title"].value_counts()
```

```python
df.groupby("snippet/title")["views"].mean().plot(kind="bar")
```

```python
df.groupby("snippet/title")["dislikes"].mean().plot(kind="bar")
```

```python
df.groupby("snippet/title")["likes"].mean().plot(kind="bar")
```

```python
df.groupby("comments_disabled")["views"].mean().plot(kind="bar")
```

```python
df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')

plt.figure(figsize=(10, 6))
plt.plot(df['trending_date'], df['views'], label='Views')
plt.plot(df['trending_date'], df['likes'], label='Likes')
plt.plot(df['trending_date'], df['comment_count'], label='Comments')
plt.xlabel('Trending Date')
plt.ylabel('Counts')
plt.title('Trends Over Time')
plt.legend()
plt.show()
```
