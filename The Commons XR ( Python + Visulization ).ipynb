```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns


plt.style.use("ggplot")
rcParams['figure.figsize'] = (12, 6)


from sklearn.datasets import load_wine
```

```python
wine = load_wine()

df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
df
```

```python
df["target"] = wine.target
```

```python
df.target.value_counts().plot(kind="bar")
plt.title("Value counts of the target variable")
plt.xlabel("Wine type")
plt.xticks(rotation=0)
plt.ylabel("Count");
```

```python
df.magnesium.describe()
```

```python
df.magnesium.hist()
plt.title("Distribution of Magnesium")
plt.xlabel("Values")
plt.xticks(rotation=0)
plt.ylabel("Frequency");
```

```python
sns.pairplot(df);
```

```python
sns.catplot(x="target", y="proline", data=df, kind="box", aspect=1.5)
plt.title("Boxplot for target vs proline");
```

```python
sns.scatterplot(x="proline", y="flavanoids", hue="target", data=df, palette="Dark2", s=80)
plt.title("Relationship between proline, flavanoids and target");
```

```python
corrmat = df.corr()
hm = sns.heatmap(corrmat, 
                 cbar=True, 
                 annot=True, 
                 square=True, 
                 fmt='.2f', 
                 annot_kws={'size': 10}, 
                 yticklabels=df.columns, 
                 xticklabels=df.columns, 
                 cmap="Spectral_r");
```
